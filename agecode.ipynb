```python
import pandas as pd
import re


myfile = pd.read_csv('C:/Users/priya/Desktop/python/Drugs-Analysis-by-Web-Scraping-master1/Drugs-Analysis-by-Web-Scraping-master/sample.csv', sep=',')


data = myfile['Reviewer Details'].tolist()
age = []
#print(data)

for d in data:
    #print(d)
    regex = r"\d\d-\d\d"
    searchObj = re.search( regex, d, re.M|re.I)
    if searchObj:
        #print ("search --> searchObj.group() : ", searchObj.group())
        age.append(searchObj.group())
    else:
        #print ("Nothing found!!")
        age.append("-")

myfile['Age']=age
print(myfile)

myfile.to_csv(r'new_sample.csv')      #Writting to a csv file to get a sample file
```
